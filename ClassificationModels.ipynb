```python
from sklearn.feature_extraction.text import TfidfVectorizer
def text_representation(data):
    tfidf_vect = TfidfVectorizer()
    data["body_en"] = data["body_en"].apply(lambda text: " ".join(set(text)))
    X_tfidf = tfidf_vect.fit_transform(data['body_en'])
    #print(X_tfidf.shape)
    #print(tfidf_vect.get_feature_names())
    X_tfidf = pd.DataFrame(X_tfidf.toarray())
    return X_tfidf
#apply the TFIDV function
X_tfidf = text_representation(X)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, target_train, target_test = train_test_split(X_tfidf, Y, test_size = 0.30, random_state = 10)
```

```python
xg = XGBClassifier(max_depth=6, n_estimators = 100, eval_metric="mlogloss")
```

```python
xgb_model = xg.fit(features_train, target_train)
```

```python
from sklearn.metrics import classification_report
y_pred = xgb_model.predict(features_test)
print(classification_report(target_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
#define cross-validation method to use
cv = KFold(n_splits=10, random_state=None, shuffle=False) 

#build multiple linear regression model
dct = DecisionTreeClassifier(random_state=1)

dct_model = dct.fit(features_train, target_train)

dct_y_pred = dct.predict(features_test)
```

```python
print(classification_report(target_test, dct_y_pred))
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold

svc = LinearSVC()
cv = StratifiedKFold(n_splits=10)
svc_model = svc.fit(features_train, target_train)

svc_y_pred = svc.predict(features_test)
```

```python
print(classification_report(target_test, svc_y_pred))
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='adam', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(features_train, target_train)
```

```python
mlp_pred = clf.predict(features_test)
print(classification_report(target_test, mlp_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gy_pred = gnb.fit(features_train, target_train).predict(features_test)
print(classification_report(target_test, gy_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=4)
neigh.fit(features_train, target_train)
kn_pred = neigh.predict(features_test)
print(classification_report(target_test, kn_pred))
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss="hinge", penalty="l2", max_iter=5)
clf.fit(features_train, target_train)
sg_pred = clf.predict(features_test)
print(classification_report(target_test, sg_pred))
```
