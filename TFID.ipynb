```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np

# Additional modules
import sys

# Data visualisation 
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_excel('english_scored_body_only.xlsx')
```

```python
from nltk.tokenize import TreebankWordTokenizer
tk2 = TreebankWordTokenizer()
tokens = []
for i in df["body"]:
    value = tk2.tokenize(i)
    #print(value)
    for j in value:
        if j not in tokens:
            tokens.append(j)
```

```python
len(tokens)
```

```python
encode = {'.':0}

for i in range(len(tokens)):
    encode[tokens[i]] = i+1
```

```python
encode
```

```python
def code(words):
    t=[]
    
    for i in words:
        t.append(encode[i])
    return t
```

```python
maxlen = 4
step = 3
sentences = []
next_chars = []

for i in range(0, len(tokens) - maxlen, step):
    sentences.append(code(tokens[i: i + maxlen]))
    next_chars.append(encode[tokens[i + maxlen]])
print('nb sequences: ', sentences[1:10])
print('nb sequences: ', next_chars[1:10])
```
